```python

#  合并表格

df_mx = pd.DataFrame()
df_cwmx = pd.DataFrame()
patt = "(?!<$)(?P<area>.+)签约明细.+\.xls[xm]?"
com = re.compile(patt)

file_dic = {}
for i,j,k in os.walk(path):
    for x in k:
        z = com.match(x)
        if '~' not in x and '河西未转'not in x and z: 
            area = z.group('area')
            try:
                print('读取' + area + '商业签约明细')
                
                dfsy = pd.read_excel(path + "\/" + x,sheetname = area+"商业签约明细")
                
                dfsy['工作簿名'] = area
                dfsy['类型'] = '商业'
                print('合并' + area + '商业签约明细')
                df_mx = pd.concat([df_mx, dfsy]) 
            except Exception as e:
                print('->无'+ area + '商业签约明细表')

            try:
                print('读取' + area + '签约明细')
                dfzf = pd.read_excel(path + "\/" + x,sheetname = area+"签约明细")
               
                
                dfzf['工作簿名'] = area
                dfzf['类型'] = '住宅'

                print('合并' + area + '签约明细')
                df_mx = pd.concat([df_mx, dfzf])
            except Exception as e:
                print('--->无'+ area + '签约明细表')
                
            try:
                print('读取' + area + '车位明细')
                dfcw = pd.read_excel(path + "\/" + x,sheetname = area+"车位明细")
               
                
                dfcw['工作簿名'] = area
                dfcw['项目'] = area
                dfcw['类型'] = '住宅'
                print('合并' + area + '车位明细')
                df_cwmx = pd.concat([df_cwmx, dfcw])
            except Exception as e:
                print('--->无'+ area + '车位明细表')


            print()
    print('end')
            
df_mx.reset_index(drop=True,inplace=True)

# 分离字段
s = df_mx['项目名称']
dd = s.str.extract(r"\s*(?P<项目名称_trim>(?P<项目>.+?)(?P<项目期数>[一二三四五六七八九]期(?:商业)?))\s*")
df_mx1 = pd.concat([dd, df_mx],axis=1)
#  检查 项目名称 这一列时候彻底分裂
df_mx1[df_mx1['项目名称_trim']!= df_mx1['项目名称']][['工作簿名','项目名称_trim','项目','项目期数','项目名称']]

#  去除无效字段
df_mx1.dropna(subset=['项目名称'],inplace=True)


# 改回名字
del df_mx1['项目名称']
df_mx1.rename(columns={'项目名称_trim':'项目名称'},inplace=True)
```

```python
"""
字段整理
"""

f_date = datetime.datetime(2017,3,17)
df_mx1['备案签字日期'] = df_mx1['备案异常情况'].str.extract(r'(20\d{2}/\d{1,2}/\d{1,2})(?=已签字|已网签)',expand=False).astype('datetime64')
df_mx1['网签日期'] = df_mx1['单机改网签日期']
df_mx1['网签日期'][(df_mx1['类型'] == "住宅") & (df_mx1['签约日期'] < f_date)] = \
    df_mx1['签约日期'][(df_mx1['类型'] == "住宅") & (df_mx1['签约日期'] < f_date)]
df_mx1['网签日期'][df_mx1['类型'] == "商业"  ] = df_mx1['签约日期'][df_mx1['类型'] == "商业"]


```

```python
def _hz(data,title):

    data = data.groupby('项目').agg({'买卖合同总价':{"套数":'count',"金额(万元)":'sum'}})
    data.columns= pd.MultiIndex.from_product( [[title],['套数','金额(万元)']])
    return data


# 在同一月份 不同天数 的判断
def _before_data(data,target_date,field):
    if target_date.year == CURRENT_DATE.year and\
       target_date.month == CURRENT_DATE.month and \
       target_date != CURRENT_DATE:
            data = data[
                data[field].dt <= target_date
            ]
    return data
    



# 给出日期属于 第几个 七天   最多 4个 过 4 为 4
def get_7days(date):
    zs = date.day//7
    ys = date.day % 7
    if ys == 0:
        zs = zs - 1
    start_day = zs * 7 + 1
    end_day =  (zs + 1) * 7 if zs < 3 else date.daysinmonth
    return start_day,end_day
```

```python


"""
签约
"""
# 本月
def qianyue_hz(data,target_date,title):

    df_t = data[['项目','买卖合同总价','签约日期']]

    df_t = df_t[
        ((df_t['签约日期'].dt.year == target_date.year) &\
        (df_t['签约日期'].dt.month == target_date.month) ) 
        ]
    df_t = _before_data(df_t,target_date,"签约日期")
    return _hz(df_t,title)

# 今日
def today_qianyue_hz(data,target_date,title):

    df_t = data[['项目','买卖合同总价','签约日期']]

    df_t = df_t[
        df_t['签约日期'].dt.date == target_date
        ]

    return _hz(df_t,title)

"""

打网签
"""
def wangqian_hz(data,target_date,title):
    df_t = data[['项目','买卖合同总价','网签日期']]
    df_t = df_t[
         (df_t['网签日期'].dt.year == target_date.year) &\
        (df_t['网签日期'].dt.month == target_date.month) 

        ]
    df_t = _before_data(df_t,target_date,"网签日期")
    return _hz(df_t,title)
def today_wangqian_hz(data,target_date,title):
    df_t = data[['项目','买卖合同总价','网签日期']]
    df_t = df_t[
         df_t['网签日期'].dt.date == target_date

        ]
    return _hz(df_t,title)


"""
备案时间
"""
def beian_hz(data,target_date,title):
    df_t = data[['项目','买卖合同总价','送备案时间']]

    df_t = df_t[

        (df_t['送备案时间'].dt.year == target_date.year) &\
        (df_t['送备案时间'].dt.month == target_date.month) 

        ]
    df_t = _before_data(df_t,target_date,"送备案时间")
    return _hz(df_t,title)



"""
本月草签+网签

"""
def cq_wq_hz(data,target_date,title):
    df_t = data[['项目','买卖合同总价','签约日期', '单机改网签日期']]
    df_t = df_t[
        ((df_t['签约日期'].dt.year == target_date.year) &\
        (df_t['签约日期'].dt.month == target_date.month) ) |\

        ((df_t['单机改网签日期'].dt.year == target_date.year) &\
        (df_t['单机改网签日期'].dt.month == target_date.month) )

        ]
    df_t = _before_data(df_t,target_date,"签约日期")
    df_t = _before_data(df_t,target_date,"单机改网签日期")
    return _hz(df_t,title)


"""
上月预告

"""
def yugao_hz(data,target_date,title):
    df_t = data[['项目','买卖合同总价', '备案完成时间']]
    df_t = df_t[
        ((df_t['备案完成时间'].dt.year == target_date.year) &\
        (df_t['备案完成时间'].dt.month == target_date.month) )
        ]

    return _hz(df_t,title)


"""
已通过网签未签字正式合同

"""
def wq_weiqianzi(data,title):
    df_t = data[['项目','买卖合同总价','备案异常情况']]

    df_t = df_t[

        df_t['备案异常情况'].astype('str').str.contains('未签字|未网签')

        ]

    return _hz(df_t,title)


"""
数据通报

"""
# 备案统计
def beian(df,date = None,interval=None):
    dfba = df[['项目','送备案时间','买卖合同总价']]
    today = date or datetime.date.today()
    tj_l = []

    # 当日

    dfqy_today = dfba[(dfba['送备案时间'] == today)]
    
    tj_l.append(_hz(dfqy_today,'今日备案'))

    

    # 获取起始段
    interval = interval or get_7days(today)
    start_day = interval[0]
    end_day = interval[1]
    
    dfqy_interval = dfba[(dfba['送备案时间'].dt.month == today.month) & \
                        (dfba['送备案时间'].dt.year == today.year) & \
                        (dfba['送备案时间'].dt.day >= start_day) &\
                        (dfba['送备案时间'].dt.day <= end_day)
                       ]
    
    tj_l.append(_hz(dfqy_interval,'%s-%s备案' %(start_day,end_day)))
    
    
    

#     当月

    
    dfqy_month = dfba[(dfba['送备案时间'].dt.month == today.month) & (dfba['送备案时间'].dt.year == today.year)]
    dfqy_month = _before_data(dfqy_month,today,"送备案时间")

    tj_l.append(_hz(dfqy_month,'本月备案'))
    return tj_l



# 判断网签  
# 字段 ：'客户契税/维修基金','备案异常情况','买卖合同总价'
def wq(df,date = None,interval=None):
    data = df[['项目','买卖合同总价', '客户契税/维修基金','类型','备案签字日期', '网签日期']]

    tj_l = []
    today = date or datetime.date.today()

    
    data = data.dropna(subset=['备案签字日期','网签日期', '客户契税/维修基金']).reset_index()

    df_t = data[['备案签字日期','网签日期', '客户契税/维修基金']]


    data['已完成网签日期'] = df_t.apply(lambda x:max(x),axis=1)


    # 当日
    td = pd.Timestamp(today)
    data_day = data[
        (data['已完成网签日期'] == today)

        ]
    
    tj_l.append(_hz(data_day,'当日完成网签' ))
    
    # 时间段
    interval = interval or get_7days(today)
    start_day = interval[0]
    end_day = interval[1]
    data_interval = data[
        (data['已完成网签日期'].dt.year == today.year) & \
         (data['已完成网签日期'].dt.month == today.month) & \
        (data['已完成网签日期'].dt.day >= start_day) & \
        (data['已完成网签日期'].dt.day <= end_day)
        ]
    
    tj_l.append(_hz(data_interval,'%s-%s完成网签' %(start_day,end_day)))

    
    # 当月
    
    data_month = data[
        (data['已完成网签日期'].dt.year == today.year) & \
         (data['已完成网签日期'].dt.month == today.month) 
        ]
    data_month = _before_data(data_month,today,"已完成网签日期")
    tj_l.append(_hz(data_month,'当月完成网签'))
   
    return tj_l



"""
合并所需 的子项汇总

"""
def hb_hz(hz_ls,index_name,index_l=None):
    df_hz = pd.concat(hz_ls,axis=1)
    df_hz.index.name = index_name

    if index_l:
        df_hz = df_hz.loc[index_l].fillna(0)
    df_hz.loc['总和'] = df_hz.apply(lambda x:sum(x))

    for lv1_col in df_hz.columns.levels[0]:
        df_hz[lv1_col,df_hz.columns.levels[1][1]] = np.around(df_hz[lv1_col,df_hz.columns.levels[1][1]]/10000)
    return df_hz
```

```python

# 本月
month_hz = []
print(month_hz)
month_hz.append(qianyue_hz(df_mx1,today,'本月签约'))
month_hz.append(wangqian_hz(df_mx1,today,'本月网签'))
month_hz.append(beian_hz(df_mx1,today,'本月备案'))
month_hz.append(cq_wq_hz(df_mx1,today,'本月草签+网签'))
month_hz.append(yugao_hz(df_mx1,today,'本月预告'))

# 上月
last_month_hz = []
last_month_hz.append(qianyue_hz(df_mx1,before_month_first,'上月签约'))
last_month_hz.append(wangqian_hz(df_mx1,before_month_first,'上月网签'))
last_month_hz.append(beian_hz(df_mx1,before_month_first,'上月备案'))
last_month_hz.append(cq_wq_hz(df_mx1,before_month_first,'上月草签+网签'))
last_month_hz.append(yugao_hz(df_mx1,before_month_first,'上月预告'))


# 数据通报
sjtb_hz = []
sjtb_hz.extend(beian(df_mx1,today,interval))
sjtb_hz.extend(wq(df_mx1,today,interval))

# 签约日期
qianyue_hz_l = []
qianyue_hz_l.append(today_qianyue_hz(df_mx1,today,'本日签约'))
qianyue_hz_l.append(month_hz[0])
qianyue_hz_l.append(today_wangqian_hz(df_mx1,today,'本日网签'))
qianyue_hz_l.append(month_hz[1])
qianyue_hz_l.append(wq_weiqianzi(df_mx1,'已通过网签未签字正式合同'))
```
