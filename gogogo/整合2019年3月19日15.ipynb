```python
# 检测 时间异常函数
def check_timefield(data, check_col, show_str_unique=False):
    filed_l = ['项目名称', '项目', '项目期数','楼栋','房号']
    
    filed_l.append(check_col)
    print(filed_l)
    df_jc = data[filed_l]
    df_jc['房序号'] = df_jc['楼栋'].astype('str') + '-' + df_jc['房号'].astype('str')
    field_unique = [x for x in df_jc[check_col].unique() if type(x) == str]
    if show_str_unique:
        print(field_unique)
    return df_jc[df_jc[check_col].isin(field_unique)]

# 检测为空字段
def check_space(data,check_col):
    filed_l = ['项目名称', '项目', '项目期数','楼栋','房号']
    filed_l.append(check_col)
    df_jc = data[filed_l]
    return df_jc[df_jc[check_col].isnull()]


```

```python
# 提取日期的最后一天
def get_lastday(td = datetime.date.today()):

    next_month = td.month + 1 if td.month <12 else 1
    next_year = td.year if td.month < 12 else td.year + 1
    last_day = (datetime.datetime(next_year,next_month,1) - datetime.datetime(td.year,td.month,1)).days

    return last_day

# 给出日期属于 第几个 七天   最多 4个 过 4 为 4
def get_7days(date):
    zs = date.day//7
    ys = date.day % 7
    if ys == 0:
        zs = zs - 1
    start_day = zs * 7 + 1
    end_day =  (zs + 1) * 7 if zs < 3 else get_lastday(today)
    return start_day,end_day

```

```python
# 备案统计
def beian(df,date = datetime.date.today(),quantum=None):
    dfba = df[['项目','送备案时间','买卖合同总价']]
    today = datetime.date.today()
    field_l = ['金域国际','金域滨江','紫台',  '金域缇香','梅溪郡','白鹭郡']
    dfba_hz = pd.DataFrame(index=field_l)
    
    # 当日
    d = date
    td = pd.Timestamp(d)
    dfqy_today = dfba[(dfba['送备案时间'] == td)]
    dfqy_today = dfqy_today.groupby('项目')\
         .agg({'买卖合同总价':{'套数':'count','金额':'sum'}})
    dfqy_today.columns = ['今日_套数','今日_金额']
    dfqy_today['今日_金额'] = np.round(dfqy_today['今日_金额']/10000)
    
    

#     dfqy_today = dfqy_today.loc[field_l].fillna(0)
    
#     return dfqy_today

    
    # 时间段
    last_day = get_lastday(today)
    
    # 获取起始段
    if quantum:
        start_time = quantum[0]
        end_time = quantum[1]
    else:
        zs = today.day//7
        ys = today.day % 7
        if ys == 0:
            zs = zs - 1
        start_time = zs * 7 + 1
        end_time =  (zs + 1) * 7 if zs < 3 else last_day
    
    dfqy_quantum = dfba[(dfba['送备案时间'].dt.month == today.month) & \
                        (dfba['送备案时间'].dt.year == today.year) & \
                        (dfba['送备案时间'].dt.day >= start_time) &\
                        (dfba['送备案时间'].dt.day <= end_time)
                       ]

    dfqy_quantum = dfqy_quantum.groupby('项目')\
         .agg({'买卖合同总价':{'套数':'count','金额':'sum'}})
    
    title = ['%s-%s套数' %(start_time,end_time),'%s-%s金额' %(start_time,end_time)]
    dfqy_quantum.columns = title
    dfqy_quantum[title[1]] = np.round(dfqy_quantum[title[1]]/10000)
    
    

#     dfqy_quantum = dfqy_quantum.loc[field_l].fillna(0)
#     return dfqy_quantum
    
    
    
#     当月

    
    dfqy_month = dfba[(dfba['送备案时间'].dt.month == today.month) & (dfba['送备案时间'].dt.year == today.year)]
    dfqy_month = dfqy_month.groupby('项目')\
         .agg({'买卖合同总价':{'套数':'count','金额':'sum'}})
    dfqy_month.columns = ['本月套数','本月金额']
    dfqy_month['本月金额'] = np.round(dfqy_month['本月金额']/10000)
    
    

#     dfqy_month = dfqy_month.loc[field_l].fillna(0)
#     return dfqy_month
    
    dfba_hz = pd.concat([dfba_hz,dfqy_today,dfqy_quantum,dfqy_month],axis=1)

    dfba_hz = dfba_hz.loc[field_l].fillna(0)
    dfba_hz.index.name = '项目'
    dfba_hz.reset_index(inplace=True)
    return dfba_hz
```

```python
# 判断网签  
# 字段 ：'客户契税/维修基金','备案异常情况','买卖合同总价'
def wq(df,date = None,interval=None):
    data = df[['项目','备案异常情况','买卖合同总价','单机改网签日期']]
    today = date or datetime.date.today()
    field_l = ['金域国际','金域滨江','紫台',  '金域缇香','梅溪郡','白鹭郡']
    
    data['备案签字日期'] = data['备案异常情况'].str.extract(r'(20\d{2}/\d{1,2}/\d{1,2})(?=已签字)',expand=False).astype('datetime64')
    
    
    # 当月
    # 筛选逻辑
    
    data_month = data[
        (data['单机改网签日期'].notnull()) & \
        (data['备案签字日期'].dt.year == today.year) & \
         (data['备案签字日期'].dt.month == today.month) 
        ]

    data_month = data_month.groupby('项目').agg({'买卖合同总价':{'套数':'count','金额':'sum'}})
    
    data_month.columns = ['套数','金额']
    data_month['金额'] = np.round(data_month['金额']  / 10000)
    data_month.columns = ['本月_套数','本月_金额']

    
    # 时间段
    interval = interval or get_7days(today)
    start_day = interval[0]
    end_day = interval[1]
    data_interval = data[
        (data['单机改网签日期'].notnull()) & \
        (data['备案签字日期'].dt.year == today.year) & \
         (data['备案签字日期'].dt.month == today.month) & \
        (data['备案签字日期'].dt.day >= interval[0]) & \
        (data['备案签字日期'].dt.day <= interval[1])
        ]
    data_interval = data_interval.groupby('项目').agg({'买卖合同总价':{'套数':'count','金额':'sum'}})
    
    data_interval.columns = ['套数','金额']
    data_interval['金额'] = np.round(data_interval['金额']  / 10000)
    title = ['%s.%s-%s.%s套数' %(today.month,start_day,today.month,end_day),'%s-%s金额' %(start_day,end_day)]
    data_interval.columns = title
    

    
    
    # 当日
    td = pd.Timestamp(today)
    data_day = data[
        (data['单机改网签日期'].notnull()) & \
        (data['备案签字日期'] == today)

        ]

    data_day = data_day.groupby('项目').agg({'买卖合同总价':{'套数':'count','金额':'sum'}})
    
    data_day.columns = ['套数','金额']
    data_day['金额'] = np.round(data_day['金额']  / 10000)
    data_day.columns = ['今日_套数','今日_金额']
    
    wq_hz = pd.concat([data_day,data_interval,data_month],axis=1)
    wq_hz.index.name = '项目'
    wq_hz = wq_hz.loc[field_l].reset_index().fillna(0)
    return wq_hz
```
